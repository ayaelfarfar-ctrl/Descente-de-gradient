```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("DataDG.csv", sep=";")
df.head()
```

```python
serie = df["China"].values
serie
```

Le modèle AR(1) suppose que la relation entre y(t) et y(t-1) est linéaire,
que la série est stationnaire (|a| < 1), et que les résidus constituent
un bruit blanc.

AR(1) : y_t = a · y_{t-1} + b

```python
X = serie[:-1]
Y = serie[1:]
```

```python
m = len(X)
m
```

*Hyper-paramètres*

```python
eta = 0.01
max_iter = 50000
epsilon = 1e-8
```

*Initialisation*

```python
np.random.seed(42)
a = np.random.rand() * 0.5
b = np.random.rand() * 0.5
```

*La boucle*

```python
converged = False
for it in range(max_iter):

    y_hat = a * X + b
    errors = y_hat - Y
    cost_function = (1 / (2 * m)) * np.sum(errors ** 2)

    gradient_a = (1 / m) * np.dot(X, errors)
    gradient_b = (1 / m) * np.sum(errors)
    gradient_norm = np.sqrt(gradient_a**2 + gradient_b**2)
    if gradient_norm < epsilon:
        converged = True
        break

    a = a - eta * gradient_a
    b = b - eta * gradient_b
```

```python
print(f"Résultats:")

if converged:
    print(f"\n Convergence à l'itération {it+1}")
else:
    print(f"\n Pas de convergence après {max_iter} itérations")

print(f" a = {a}")
print(f" b = {b}")
print(f" Coût final = {cost_function}")
print(f" Modèle : y(t) = {a} · y(t-1) + {b}")
```

*OLS*

```python
x_bar = np.mean(X)
y_bar = np.mean(Y)
a_ols = np.sum((X - x_bar) * (Y - y_bar)) / np.sum((X - x_bar) ** 2)
b_ols = y_bar - a_ols * x_bar
```

```python
print(f"DG  → a = {a:.6f},     b = {b:.6f}")
print(f"OLS → a = {a_ols:.6f}, b = {b_ols:.6f}")
```
